### daywise sales using line chart

```python
sales_over_time = df.groupby(['Order Date'])['Sales'].sum().reset_index()
```

```python
sales_over_time
```

```python
sns.lineplot(data=sales_over_time,x='Order Date',y='Sales')
```

```python
sns.lineplot(data=sales_over_time,x='Order Date',y='Sales',marker='o')
```

```python
plt.figure(figsize=(10,10))
sns.lineplot(data=sales_over_time,x='Order Date',y='Sales',marker='o')
```

```python
plt.figure(figsize=(10,10))
sns.lineplot(data=sales_over_time,x='Order Date',y='Sales',marker='o')
plt.title('sales over time')
plt.xlabel('Order Date')
plt.ylabel('Sales')
plt.grid()
plt.legend('Order Date')
plt.show()
```

### sale by category over time

```python
sales_by_category_over_time = df.groupby(['Order Date','Category'])['Sales'].sum().reset_index()
```

```python
plt.figure(figsize=(10,8))
sns.lineplot(data=sales_by_category_over_time,x='Order Date',y='Sales',marker='o',hue='Category')
plt.title('sales over time')
plt.xlabel('Order Dates')
plt.ylabel('Sales')
plt.grid()
plt.legend('Order Date')
plt.show()
```
